```python
import pandas as pd

# Select numerical columns only
num_df = df.select_dtypes(include=['int64', 'float64'])
n_rows = len(df)

outlier_report = []

for col in num_df.columns:
    data = num_df[col].dropna()  # ignore missing
    
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Count outliers
    outliers_low = (data < lower_bound).sum()
    outliers_high = (data > upper_bound).sum()
    total_outliers = outliers_low + outliers_high
    outlier_pct = (total_outliers / n_rows) * 100

    # Suggestion based on severity
    if outlier_pct == 0:
        suggestion = "No action needed"
    elif outlier_pct <= 5:
        suggestion = "Cap/Floor outliers"
    else:
        suggestion = "Consider transformation or removal"

    outlier_report.append({
        "Column": col,
        "Outliers_Count": total_outliers,
        "Outliers_%": round(outlier_pct, 2),
        "Suggestion": suggestion
    })

outlier_df = pd.DataFrame(outlier_report)
outlier_df
```

# Outlier Filtering Function


    Automatically detects and handles outliers in numeric columns for ML.
    
    Parameters:
    -----------
    df : pandas.DataFrame
        Input dataframe
    numeric_cols : list, optional
        List of numeric columns to check. If None, auto-detect numeric columns.
    threshold_pct : float
        % of outliers above which robust transformation is applied.
    return_summary : bool
        If True, return a summary DataFrame with outlier counts & actions.
    
    Returns:
    --------
    df_clean : pandas.DataFrame
        DataFrame with outliers handled.
    (Optional) outlier_summary_df : pandas.DataFrame
        Summary of outliers per column (only if return_summary=True)

```python

from sklearn.preprocessing import PowerTransformer

def handle_outliers_ml(df, numeric_cols=None, threshold_pct=5, return_summary=False):

    
    df_clean = df.copy()
    
    if numeric_cols is None:
        numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
    
    n_rows = len(df)
    outlier_summary = []

    for col in numeric_cols:
        data = df_clean[col].dropna()
        Q1 = data.quantile(0.25)
        Q3 = data.quantile(0.75)
        IQR = Q3 - Q1

        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR

        outliers_low = (data < lower).sum()
        outliers_high = (data > upper).sum()
        total_outliers = outliers_low + outliers_high
        outlier_pct = (total_outliers / n_rows) * 100

        # Decide method
        if total_outliers == 0:
            method = "No action"
        elif outlier_pct <= threshold_pct:
            df_clean[col] = df_clean[col].clip(lower, upper)
            method = "Cap/Floor"
        else:
            pt = PowerTransformer(method='yeo-johnson')
            df_clean[col] = pt.fit_transform(df_clean[[col]])
            method = "Power Transform (Yeo-Johnson)"

        # Append summary if needed
        if return_summary:
            outlier_summary.append({
                "Column": col,
                "Outliers_Count": total_outliers,
                "Outliers_%": round(outlier_pct, 2),
                "Action": method
            })

    if return_summary:
        outlier_summary_df = pd.DataFrame(outlier_summary)
        return df_clean, outlier_summary_df
    else:
        return df_clean
```

# desgining a function to handel missing values
    """
    Handles missing values in a DataFrame.
    - Numeric columns: fill with median
    - Categorical columns: fill with most frequent value
    Returns a null-free DataFrame.
    
    Parameters:
    - df : pd.DataFrame
    - numeric_cols : list of numeric columns (auto-detect if None)
    - categorical_cols : list of categorical columns (auto-detect if None)
    
    Returns:
    - df_clean : pd.DataFrame (null-free)
    """


```python

from sklearn.impute import SimpleImputer

# Define imputers
num_imputer = SimpleImputer(strategy='median')
cat_imputer = SimpleImputer(strategy='most_frequent')

def handle_missing_values(df, numeric_cols=None, categorical_cols=None):
    df_clean = df.copy()
    
    # Auto-detect columns if not provided
    if numeric_cols is None:
        numeric_cols = df_clean.select_dtypes(include=['int64', 'float64']).columns.tolist()
    if categorical_cols is None:
        categorical_cols = df_clean.select_dtypes(include=['object']).columns.tolist()
    
    # Fill numeric missing values
    if numeric_cols:
        df_clean[numeric_cols] = num_imputer.fit_transform(df_clean[numeric_cols])
    
    # Fill categorical missing values
    if categorical_cols:
        df_clean[categorical_cols] = cat_imputer.fit_transform(df_clean[categorical_cols])
    
    return df_clean
```

# now outliners 
    """
    Handles outliers strictly by replacing only values outside IQR limits.
    - Leaves all other values unchanged
    - Automatically skips binary columns (0/1)
    
    Parameters:
    - df : pd.DataFrame
    - numeric_cols : list of numeric columns (auto-detect if None)
    - return_summary : bool, if True returns summary DataFrame
    
    Returns:
    - df_clean : DataFrame with outliers replaced
    - optional: summary DataFrame if return_summary=True
    """

```python
import pandas as pd

def handle_outliers_strict(df, numeric_cols=None, return_summary=False):

    df_clean = df.copy()
    
    # Auto-detect numeric columns if not provided
    if numeric_cols is None:
        numeric_cols = df_clean.select_dtypes(include=['int64', 'float64']).columns.tolist()
    
    # Detect binary columns (0/1) and skip them
    binary_cols = [col for col in numeric_cols if df_clean[col].nunique() == 2]
    numeric_cols_to_handle = [col for col in numeric_cols if col not in binary_cols]
    
    n_rows = len(df_clean)
    summary = []
    
    for col in numeric_cols_to_handle:
        data = df_clean[col]
        Q1 = data.quantile(0.25)
        Q3 = data.quantile(0.75)
        IQR = Q3 - Q1
        
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        
        # Vectorized replacement for outliers
        df_clean[col] = data.mask(data < lower, lower)
        df_clean[col] = df_clean[col].mask(df_clean[col] > upper, upper)
        
        if return_summary:
            outliers_low = (data < lower).sum()
            outliers_high = (data > upper).sum()
            total_outliers = outliers_low + outliers_high
            summary.append({
                "Column": col,
                "Outliers_Count": total_outliers,
                "Lower_Threshold": lower,
                "Upper_Threshold": upper
            })
    
    if return_summary:
        return df_clean, pd.DataFrame(summary)
    else:
        return df_clean
```

# now making a column transformer 

```python
numericals_list=['ProdTaken',
 'Age',
 'CityTier',
 'DurationOfPitch',
 'NumberOfPersonVisiting',
 'NumberOfFollowups',
 'PreferredPropertyStar',
 'NumberOfTrips',
 'Passport',
 'PitchSatisfactionScore',
 'OwnCar',
 'NumberOfChildrenVisiting',
 'MonthlyIncome']
```

```python

from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder, OrdinalEncoder
from sklearn.impute import SimpleImputer


# 1ï¸âƒ£ Define imputers

num_imputer = SimpleImputer(strategy='median')
cat_imputer = SimpleImputer(strategy='most_frequent')

# 2ï¸âƒ£ Placeholder column lists

numeric_cols = numericals_list  # replace with your numeric column list
ordinal_cols = list(ordinal_dict.keys())  # from your ordinal dict
ordinal_categories = list(ordinal_dict.values())
ohe_cols = list(ohe_dict.keys())  # from your OHE dict
ohe_categories = list(ohe_dict.values())

# 3ï¸âƒ£ ColumnTransformer shell

preprocessor = ColumnTransformer(
    transformers=[
        ('num', Pipeline([
            ('imputer', num_imputer),
            ('scaler', StandardScaler())
        ]), numeric_cols),
        
        ('ord', Pipeline([
            ('imputer', cat_imputer),
            ('ordinal', OrdinalEncoder(categories=ordinal_categories))
        ]), ordinal_cols),
        
        ('ohe', Pipeline([
            ('imputer', cat_imputer),
            ('ohe', OneHotEncoder(sparse_output=False, drop='first', categories=ohe_categories))
        ]), ohe_cols)
    ],
    remainder='drop'  # drop any other columns not listed
)
```

# now making pipeline 

```python
from sklearn.ensemble import RandomForestClassifier
pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('model', RandomForestClassifier(n_estimators=200, random_state=42))
])
```

# finding best parameters 

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'model__n_estimators': [100, 200],
    'model__max_depth': [None, 10, 20],
    'model__min_samples_split': [2, 5],
    'model__min_samples_leaf': [1, 2],
    'preprocessor__num__imputer__strategy': ['median', 'mean'],
    'preprocessor__ord__imputer__strategy': ['most_frequent'],
    'preprocessor__ohe__imputer__strategy': ['most_frequent']
}

grid_search = GridSearchCV(
    estimator=pipeline,
    param_grid=param_grid,
    cv=5,
    n_jobs=-1,
    scoring='accuracy',
    verbose=2
)
```
